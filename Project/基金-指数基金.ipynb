```python
#指数涨幅分析
def fund_market_analysis(fund_list,data_root='./'):
    market_results=[]
    start_date=min_date#pd.datetime.today().strftime('%Y')+'-01-01'
    end_date=pd.datetime.today()-timedelta(1)
    #行业类别代码    
    for code in fund_list:
        #指数名称
        fund_name=fund_all_list[code] 
        #显示进度
        print ('行情统计:', fund_name,end='\r')
        #获取数据
        df=get_price(code,start_date=start_date,end_date=end_date,frequency='daily', 
                      fields=['open','close','low','high','volume','money','pre_close']) 
        
        if len(df)==0 :
            continue
        #1日涨幅
        rise_1d=(df['close'].iloc[-1]-df['pre_close'].iloc[-1])/df['pre_close'].iloc[-1]*100.0
        #5日涨幅
        rise_5d=(df['close'].iloc[-1]-df['pre_close'].iloc[-5])/df['pre_close'].iloc[-5]*100.0
        #10日涨幅
        rise_10d=(df['close'].iloc[-1]-df['pre_close'].iloc[-10])/df['pre_close'].iloc[-10]*100.0
        #20日涨幅
        rise_20d=(df['close'].iloc[-1]-df['pre_close'].iloc[-20])/df['pre_close'].iloc[-20]*100.0
        #年初至今涨幅
        rise_Early=(df['close'].iloc[-1]-df['pre_close'].iloc[0])/df['pre_close'].iloc[0]*100.0
        #30日涨幅
        rise_30d=(df['close'].iloc[-1]-df['pre_close'].iloc[-30])/df['pre_close'].iloc[-30]*100.0
        #90日涨幅
        rise_90d=(df['close'].iloc[-1]-df['pre_close'].iloc[-90])/df['pre_close'].iloc[-90]*100.0
        #250日涨幅
        rise_250d=(df['close'].iloc[-1]-df['pre_close'].iloc[-250])/df['pre_close'].iloc[-250]*100.0
        market_results.append([fund_name,
                               round(rise_1d,3),round(rise_5d,3),round(rise_10d,3),round(rise_20d,3),
                               round(rise_Early,3), 
                               round(rise_30d,3), round(rise_90d,3), round(rise_250d,3), 
                           ])  
        
    market_columns=['指数','1日','5日','10日','20日','年初至今','30日','90日','250日']
    df=pd.DataFrame(data=market_results,index=fund_list.keys(),columns=market_columns)  
    #去除代码后缀
    df.index=df.index.astype('str').str[0:6] 
    #按照涨幅排序
    return df
```
