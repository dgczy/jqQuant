## 1.准备

```python
#导入库
import numpy as np
import pandas as pd

import sys
sys.path.append('/home/jquser')

from pf_plt import Plt

plt=Plt('csv','../Data/')
```

## 2.获取数据

```python
plt.data.update(plt.pool.track)
```

## 3.数据分析

```python
plt.value.update(plt.pool.track,[10,5,3,0],True)
plt.change.update(plt.pool.track,['dyr','pe','pb'],force=True)
```

## 4.估值表格

### 4.1.PE

```python
plt.value.table.pe()
```

### 4.2.PB

```python
plt.value.table.pb()
```

### 4.3.股息率

```python
plt.value.table.dyr()
```
